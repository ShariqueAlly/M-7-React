#Q-1

 Redux is a state management library for JavaScript apps that helps store and manage the global state of THE application in a predictable and organized way.
  
 Why is Redux used in web application development?

Redux is used because it:

Manages global state in one place (Store)

Ensures predictable behavior using pure reducer functions

Makes debugging easier with tools like Redux DevTools

Helps share data across many components without prop-drilling

Works well with React and large, complex applications

#Q-2

Core Concepts of Redux
1. Store

The single source of truth.

Holds the entire application's global state.

2. Actions

Plain JavaScript objects that describe what happened.

Must have a type field.

Example: { type: "ADD_TODO", payload: "Buy milk" }

3. Reducers

Pure functions that take (state, action) and return a new state.

They decide how the state changes based on the action.

4. Dispatch

The method used to send actions to the store.

Example: store.dispatch({ type: "ADD_TODO" })

5. Subscribe

Allows components to listen for store changes and update the UI.

Example: store.subscribe(() => console.log("State changed"))

6. State

The actual data managed by Redux.

Read-only — can only be changed by dispatching actions.

#Q-3

Redux Thunk is a middleware that allows  to write asynchronous logic (like API calls) inside Redux by letting you dispatch functions instead of only plain action objects.

 How does it help with async logic?

Without Thunk → Redux can only handle synchronous actions.

With Thunk → We can:

Delay a dispatch until an async task finishes

Dispatch multiple actions (e.g., loading, success, error)

 async functions like:

const fetchUsers = () => {
  return async (dispatch) => {
    dispatch({ type: "FETCH_USERS_START" });
    const res = await fetch("/users");
    dispatch({ type: "FETCH_USERS_SUCCESS", payload: res });
  };
};


#Q-4

Benefits of Using Redux Toolkit (RTK)

Less boilerplate
RTK reduces the long, repetitive code needed in classic Redux.

createSlice simplifies reducers & actions
Automatically generates action creators and reducers in one place.

Immutability handled automatically
Uses Immer internally, so you can write “mutable-looking” code safely.

Built-in support for async logic
createAsyncThunk makes API calls easy and structured.

Better project structure
Encourages a clean, scalable pattern for large applications.

configureStore() improves setup
Comes with recommended middleware (like Thunk) and DevTools enabled by default.

Performance improvements
Avoids common Redux mistakes and ensures efficient state updates.

Official, modern standard for Redux
RTK is now the official way to write Redux apps.

#Q-5

import { createSlice } from "@reduxjs/toolkit";

const userSlice = createSlice({
  name: "user",
  initialState: {
    name: "Guest",
    isLoggedIn: false
  },
  reducers: {
    LOGIN: (state) => {
      state.isLoggedIn = true;  // RTK uses Immer → allows "mutable" update
    }
  }
});

export const { LOGIN } = userSlice.actions;
export default userSlice.reducer;
